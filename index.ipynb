```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = './driver_income.json'
data = pd.read_json(file_path)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(18, 15))
```

```python
plt.subplot(3, 1, 1)
sns.countplot(data=data, x='tenure', hue='tenure', palette='Blues', dodge=False)
plt.title('Distribution of Vehicle Tenure')
plt.xlabel('Tenure (Years)')
plt.ylabel('Count of Vehicles')
plt.legend([],[], frameon=False)
```

```python
plt.subplot(3, 1, 2)
sns.boxplot(data=data, x='ownership', y='income', color='lightgreen')
plt.title('Income Distribution by Ownership Status')
plt.xlabel('Ownership Status')
plt.ylabel('Income')
```

```python
plt.subplot(3, 1, 3)
sns.scatterplot(data=data, x='avg_distance_normalized', y='income', hue='ownership', palette='viridis', alpha=0.7)
plt.title('Income vs. Normalized Distance Traveled')
plt.xlabel('Normalized Distance Traveled')
plt.ylabel('Income')
```
